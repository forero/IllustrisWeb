```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import scipy.ndimage.filters
import numpy.linalg
%matplotlib inline
```

```python
filename = 'data/features_prediction_Tree.hdf5'
f = h5py.File(filename, 'r')
print(f.keys())
data = {}
for k in f.keys():
    data[k] = f[k][...]
f.close()

def plot_xy(data, x, y, ii):
    plt.scatter(data[x], data[y], c=data['g_frac']*data['f1_av'], s=150, alpha=0.3)
    plt.scatter(data[x][ii], data[y][ii], color='red')
    plt.xlabel(x)
    plt.ylabel(y)
```

```python
plt.figure(figsize=(8,8))

ii = data['f1_av']>0.59

plt.subplot(2,2,1)
plot_xy(data, 'f1_av', 'g_frac', ii)

plt.subplot(2,2,3)
plot_xy(data, 'ra_cut', 'smooth', ii)

plt.subplot(2,2,2)
plot_xy(data, 'max_depth', 'lambda_th', ii)

plt.subplot(2,2,4)
plot_xy(data, 'max_depth', 'smooth', ii)
```

```python
plt.figure(figsize=(8,8))

ii = data['g_frac']>0.65

plt.subplot(2,2,1)
plot_xy(data, 'f1_av', 'g_frac', ii)

plt.subplot(2,2,3)
plot_xy(data, 'ra_cut', 'smooth', ii)

plt.subplot(2,2,2)
plot_xy(data, 'max_depth', 'lambda_th', ii)

plt.subplot(2,2,4)
plot_xy(data, 'max_depth', 'smooth', ii)
```

```python
plt.figure(figsize=(8,8))

ii = (data['g_frac']>0.67) & (data['f1_av']>0.53)

plt.subplot(2,2,1)
plot_xy(data, 'f1_av', 'g_frac', ii)

plt.subplot(2,2,3)
plot_xy(data, 'ra_cut', 'smooth', ii)

plt.subplot(2,2,2)
plot_xy(data, 'max_depth', 'lambda_th', ii)

plt.subplot(2,2,4)
plot_xy(data, 'max_depth', 'smooth', ii)
```

```python
plt.figure(figsize=(8,8))

ii = (data['lambda_th']==0.3) & (np.int_(data['max_depth'])==11) & (data['smooth']==1.0)

plt.subplot(2,2,1)
plot_xy(data, 'f1_av', 'g_frac', ii)

plt.subplot(2,2,3)
plot_xy(data, 'ra_cut', 'smooth', ii)

plt.subplot(2,2,2)
plot_xy(data, 'max_depth', 'lambda_th', ii)

plt.subplot(2,2,4)
plot_xy(data, 'max_depth', 'smooth', ii)
```
